```python
# Use this to read the data for the first time

import os
import pandas as pd

def load_building_data(root_directory):
    """
    Loads all CSV files from a hierarchical dataset structure.
    
    Args:
        root_directory (str): Path to the root directory of the dataset.
    
    Returns:
        dict: Nested dictionary with data categorized by building, category, subfolder, and file type.
    """
    data = {}  # To store all the data

    # Traverse each building
    for building in os.listdir(root_directory):
        building_path = os.path.join(root_directory, building)
        if os.path.isdir(building_path):  # Ensure it's a directory
            data[building] = {}
            
            # Traverse each category (e.g., AHU, Building Total Mains)
            for category in os.listdir(building_path):
                category_path = os.path.join(building_path, category)
                if os.path.isdir(category_path):
                    data[building][category] = {}
                    
                    # Traverse each subfolder (e.g., 0, 1, 2)
                    for subfolder in os.listdir(category_path):
                        subfolder_path = os.path.join(category_path, subfolder)
                        if os.path.isdir(subfolder_path):
                            data[building][category][subfolder] = {}
                            
                            # Load each CSV file in the subfolder
                            for file in os.listdir(subfolder_path):
                                if file.endswith(".csv"):
                                    file_path = os.path.join(subfolder_path, file)
                                    file_type = file.split(".")[0]  # Extract file type (e.g., Current, Energy, Power)
                                    
                                    # Read the CSV into a DataFrame
                                    data[building][category][subfolder][file_type] = pd.read_csv(file_path)
    
    return data

# Path to the dataset root directory
root_dir = "iiitd"

# Load the data
dataset = load_building_data(root_dir)

```

```python
# Use this if the data is already read and pre-saved

import pickle

with open("iiitd/data.pkl", "wb") as file:  # Use 'wb' for write-binary mode
    pickle.dump(dataset, file)

print("Data saved to data.pkl")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt


# Ensure all data have the same length
assert len(academic_ahu_current) == len(academic_ahu_energy) == len(academic_ahu_power), "Datasets must have the same length"

# Combine current and power as input features, and energy as the target
data = pd.concat([
    academic_ahu_current['value'].rename('current'),
    academic_ahu_power['value'].rename('power'),
    academic_ahu_energy['value'].rename('energy')
], axis=1)

# Handle missing values if any
data.dropna(inplace=True)

# Scale the data using MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data, columns=data.columns)

# Create sequences for time-series data
def create_sequences(data, feature_columns, target_column, sequence_length=10):
    """
    Create sequences for time series data.

    Args:
        data (DataFrame): The dataset.
        feature_columns (list): Columns to use as input features.
        target_column (str): The column name for the target variable.
        sequence_length (int): Number of past timesteps to include in each sequence.

    Returns:
        X (ndarray): Features.
        y (ndarray): Targets.
    """
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data.iloc[i:i+sequence_length][feature_columns].values)
        y.append(data.iloc[i+sequence_length][target_column])
    return np.array(X), np.array(y)

# Define input features and target
feature_columns = ['current', 'power', 'energy']  # Exclude 'energy' from input
target_column = 'energy'

# Create input features (X) and target (y)
sequence_length = 10  # Use past 10 timesteps
X, y = create_sequences(scaled_data, feature_columns=feature_columns, target_column=target_column, sequence_length=sequence_length)

# Split into training and testing sets
train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build the LSTM model
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
    LSTM(32),
    Dense(1)  # Predict energy
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=3, batch_size=32, validation_data=(X_test, y_test))

# Evaluate the model
loss, mae = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}, Test MAE: {mae}")

# Predict on test data
predicted_energy = model.predict(X_test)

# Inverse transform the predictions and actual values
predicted_energy = scaler.inverse_transform(
    np.concatenate([np.zeros((len(predicted_energy), 2)), predicted_energy], axis=1)
)[:, 2]
actual_energy = scaler.inverse_transform(
    np.concatenate([np.zeros((len(y_test), 2)), y_test.reshape(-1, 1)], axis=1)
)[:, 2]

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(actual_energy, label='Actual Energy', alpha=0.7)
plt.plot(predicted_energy, label='Predicted Energy', alpha=0.7)
plt.legend()
plt.title('Actual vs Predicted Energy')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.show()
```
